### Add data about clusters of gene expression for homeobox genes

Load annotations

```python
Pdum_homeobox_annotations = pd.read_csv\
    (pathlib.Path(r'C:\worms\new\combined_annotations\Pdum_homeobox_annotations.csv'), sep='\t')
Pdum_homeobox_annotations['clusters'] = ''
Pdum_homeobox_annotations.head()
```

```python
Pele_homeobox_annotations = pd.read_csv\
    (pathlib.Path(r'C:\worms\new\combined_annotations\Pele_homeobox_annotations.csv'), sep='\t')
Pele_homeobox_annotations['clusters'] = ''
Pele_homeobox_annotations.head()
```

Load data about gene expression clusters from Clust

```python
clusters_homeo_df = {}
p = pathlib.Path(r'C:\worms\new\combined_annotations\cluster_info').glob('**/*')
files = [x for x in p if x.is_file()]
for cluter_file in files:
    clusters_homeo_df[cluter_file.stem.rstrip('_hom')] = pd.read_csv(cluter_file, sep=',')
```

Combine expression data and annotation

```python
for species_site in clusters_homeo_df.keys():
    species, site = species_site.split('_')
    for column in clusters_homeo_df[species_site].columns:
        site_cluster = site+'_'+column
        if species == 'pdum':
            Pdum_homeobox_annotations.loc\
                [Pdum_homeobox_annotations['query'].isin(list(clusters_homeo_df[species_site][column].dropna())),'clusters'] += site_cluster
            Pdum_homeobox_annotations[site_cluster]\
                = Pdum_homeobox_annotations['query'].isin(list(clusters_homeo_df[species_site][column].dropna()))
                
        elif species == 'pel':
            Pele_homeobox_annotations.loc\
                [Pele_homeobox_annotations['query'].isin(list(clusters_homeo_df[species_site][column].dropna())),'clusters'] += site_cluster
            Pele_homeobox_annotations[site_cluster]\
                = Pele_homeobox_annotations['query'].isin(list(clusters_homeo_df[species_site][column].dropna()))
```

### Extracting orthogroups with homeobox genes of interesting from proteinortho results for further phylogenetic analysis

```python
results_tsv_file = r'C:\worms\new\proteinortho_results\summaries\12_species.proteinortho.tsv'
proteinortho_df = pd.read_csv(results_tsv_file, sep='\t')
proteinortho_df.head()
```

```python
def get_all_orthologs(gene_name_pattern):
    indices = []
    pdum_genes = Pdum_homeobox_annotations\
        [Pdum_homeobox_annotations['Preferred_name'].str.contains(gene_name_pattern,case=False)]['query'].to_list()
    for gene in pdum_genes:
        if proteinortho_df[proteinortho_df['Pdum_amin_av.fasta'].str.contains(gene)].shape[0] != 0:
            indices.append(proteinortho_df[proteinortho_df['Pdum_amin_av.fasta'].str.contains(gene)].index[0])
            print(proteinortho_df[proteinortho_df['Pdum_amin_av.fasta'].str.contains(gene)].index[0])
        else:
            print('singletone')

    pele_genes = Pele_homeobox_annotations\
        [Pele_homeobox_annotations['Preferred_name'].str.contains(gene_name_pattern,case=False)]['query'].to_list()
    for gene in pele_genes:
        if proteinortho_df[proteinortho_df['Pele_amin_av.fasta'].str.contains(gene)].shape[0] != 0:
            indices.append(proteinortho_df[proteinortho_df['Pele_amin_av.fasta'].str.contains(gene)].index[0])
            print(proteinortho_df[proteinortho_df['Pele_amin_av.fasta'].str.contains(gene)].index[0])
        else:
            print('singletone')

    return ','.join([gene for gene in np.ravel(proteinortho_df.iloc[list(set(indices)),[3,4,5,6,7,8,9,10,11,14]].values) if gene != '*']\
        + pdum_genes\
        + pele_genes)
```

Saving genes from orthogroups of interest to extract fasta sequence with proteinortho grab_proteins.pl utility

```python
genes_of_interest = ['hox.*7','pbx','nkx2']
for gene in genes_of_interest:
    with open(pathlib.Path(r'C:\worms\new\proteinortho_results\orthogroups_for_phylogeny.txt'), 'a') as orthogroups_file:
        orthogroups_file.write(gene+' '+get_all_orthologs(gene)+'\n')
```
