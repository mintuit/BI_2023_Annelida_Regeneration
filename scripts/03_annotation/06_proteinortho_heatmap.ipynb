```python
import pathlib 
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
proteinortho_n_orthologs_folder = r'C:\worms\GITHUB_final_repository\03_SF_annotation\proteinortho\heatmap_orthologs.txt'
proteinortho_adj_matrix = pd.read_csv\
    (pathlib.Path(proteinortho_n_orthologs_folder), sep='\t', index_col=0).iloc[:,:11]
```

```python
species_full_names = ['A.melifera','B.glabrata','C.teleta','C.gigas','D.girociliatus','H.sapiens',
                      'L.gigantea','O.fusiformis','P.dumerilii','P.elegans','P.canaliculata']
```

```python
proteinortho_adj_matrix.columns = species_full_names
proteinortho_adj_matrix.index = species_full_names
```

```python
proteinortho_adj_matrix
```

```python
orthologs_heatmap = sns.heatmap(proteinortho_adj_matrix)
fig = orthologs_heatmap.get_figure()
plt.tight_layout()
fig.figure.savefig("output.png")
```
